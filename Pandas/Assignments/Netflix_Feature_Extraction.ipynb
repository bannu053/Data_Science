```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("netflix_titles.csv.zip")
df.head(1)
```

```python
# 1) no of genres using feature extraction
# listed_in has multiple genres separated by commas
# 👉 Count number of genres per show

def Extration_genres(txt):
  count=1
  for i in txt:
    if(i == ","):
      count+=1
  return count

  # creating a new column for number of genres

df["no_of_genres"]=df["listed_in"].apply(Extration_genres)
df.head(5)
```

```python
# Primary Genre
# Extract first genre from listed_in

def Extraction_primary_genre(txt):
  data=""
  for i in txt:
    if(i==","):
      break;
    data+=i
  return data

# creating a new column primary genre

df["primary_genre"]=df["listed_in"].apply(Extraction_primary_genre)
df.head(5)
```

```python
# # Duration Number From duration column
# Example: "90 min" → 90
#          "2 Seasons" → 2

def Extraction_duration(txt):
  data=""
  for i in txt:
    if(i==" "):
      break
    data+=i
  return data

  # create a new column duration with data type integer

df["duration_int"]=df["duration"].apply(Extraction_duration)
df.head(3)
```

```python
# Duration Number From duration column
# Example: "90 min" → 90
#          "2 Seasons" → 2

def Extraction_duration(txt):
  txt = str(txt)
  data=""
  for i in txt:
    if(i==" "):
      break
    if(i=="n" or i=="N"):
      data="0"
    else:
     data+=i
  return data

  # create a new column duration with data type integer

df["duration_int"]=df["duration"].apply(Extraction_duration)
# now they are in string data type ; we have to change them to int
df["duration_int"]=df["duration_int"].astype(int)

df.head(3)
```

```python
# Content Type Flag
# Movie = 0, TV Show = 1
def extraction_Type_flag(txt):
  flag=""
  for i in txt:
    if(i=="M"):
      return 0
    else:
      return 1

# create a new column type flag

df["type_flag"]=df["type"].apply(extraction_Type_flag)
df.head(3)

```

```python
# Actor Count From cast column
# 👉 Count number of actors per show

def extract_cast_count(txt):
    if pd.isna(txt):
        return 0

    return len(txt.split(","))

  # create new column cast count

df["cast_count"]=df["cast"].apply(extract_cast_count)
df.head(2)

```

```python
from datetime import datetime

current_year = datetime.now().year
df["release_year"] =df["release_year"].fillna(0)
df["release_year"]=df["release_year"].astype(int)
df["content_age"] = current_year - df["release_year"]
df.head(2)
```
