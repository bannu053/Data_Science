```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('anime.csv')
df.head()
```

```python
#make a new column for episode count
def Episode_extraction(txt):
  check=False
  data=""
  for i in txt:
    if(i==')'):
      break
    if(check==True):
     data+= i
    if(i=='('):
      check=True


  return data
```

```python
df["Episodes"] = df["Title"].apply(Episode_extraction)
df.head()
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps","")
df["Episodes"] = df["Episodes"].astype(int)
df.head()
```

```python
#which anime has the highest score
df[df['Score'] == df['Score'].max()]['Title']
```

```python
#give me frequency of each score
df["Score"].value_counts().head()
```

```python
#make a new column for time stamp
def duration_extraction(txt):
  data=""
  check=False
  for i in range(len(txt)):
    if (txt[i] == ')'):
      for j in range(i+1,i+18):
        data+=txt[j]
  return data
```

```python
df["Duration"] = df["Title"].apply(duration_extraction)
df.head()
```
