```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned__netflix.csv")
df.head()
```

```python
# How many total rows and columns are there?
df.shape
```

```python
# What are the column names?
df.columns
```

```python
# How many missing values in each column?
df.isnull().sum()
```

```python
# What are the unique values in "type"?
df['type'].unique()
```

```python
# Count how many Movies vs TV Shows
df['type'].value_counts()
```

```python
# How many shows are from India?
df[df['country']=="India"]['country'].value_counts()
```

```python
# Which year has the highest number of releases?
df['release_year'].value_counts().head(1)
```

```python
# Most common rating (TV-MA, PG, etc.)
df['rating'].value_counts().head(1)
```

```python
# Find all shows that contain “Love” in title
df[df['title'].str.contains("Love")]['title']
```

```python
# Count how many titles contain numbers
df['title'].str.contains(r"\d").value_counts()
```

```python
# Find longest movie title
max_len = df['title'].str.len().max()
df[df['title'].str.len() == max_len]['title']
```
